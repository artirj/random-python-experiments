```python
import numpy as np
import pandas
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
n=100
a=np.arange(0,n)
b=2*a+np.random.normal(size=n,scale=0.1,loc=0)*[i**1.1 for i in range(n)]
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.regplot(a,b)
plt.subplot(1,2,2)
sns.residplot(a,b)
np.polyfit(a,b,1)
```

```python
sns.regplot(a,b)
sns.regplot(a/b.std(),b/b.std())
np.polyfit(a/b.std(),b/b.std(),1)
```
